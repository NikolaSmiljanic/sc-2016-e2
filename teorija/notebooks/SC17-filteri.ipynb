# Filteri

Tema ovog predavanja je obrada toka podataka. 

Tok podataka ('stream') je sekvenca podataka koji su zavisni od vremena. Obrada ovakvih podataka razlikuje se od obrade u blokovima ('batch'). Blokovska obrada podrazumevada se podaci prvo neko vreme prikupljaju pa se nakon toga obrađuju.

Obrada toka podataka kao rezultat proizvodi novi tok pa je moguće ulančavanje tokova prilikom obrade. Novi tok je vremenski 'pomeren' u odnosnu na orginalni tok kao posledica vremena obrade. 

Kod blokovske obrade postoji i kašnjene u obradi prouzrokovano čekanjem da se blok popuni pre nego što se počne sa obradom. Blokovska obrada je pogodn kod sporijih sistema.  

Veoma često se obrada toka i obrada bloka u zavisnosti od primene mešaju. Veoma bitan podatak za obradu toka podataka je vreme odabiranja ili frekvencija. Frekvencija je obrnuto proporcionalna vremenu odabiranja. 

Na primer, ako je frekvencija toka koji treba da obradimo 100Hz tada je vreme koje protekne između dva podatka (vreme odabiranja) 1/100 deo sekunde ili 10ms. Ako sa druge strane rezultat obrade treba da koristimo sa frekvencijom od 1Hz onda je moguće obradu raditi blokovski tako što bi veličina bloka bila 100 elemenata. Međutim, ako je potreba za podacima takva da izlazni tok bude 100Hz onda bi blokovskom obradom došlo do velikog kašnjenja. 

Primeri obrade tokova podataka su analiza videa u realnom vremenu ili analiza podataka o poziciji kod GPS-a...

Problem koji se dodatno pojavljuje je problem grešaka, prouzrokovan nepreciznostima u procesu merenja i prikupljanja podataka. 

Poseban izazov kod obrade realnih podataka je 'čišćenje' toka da tako da rezultujući tok u sebi sadrži manje nepreciznosti i neodređenosti.

## Usrednjavanje - 'moving average'
 - Jednostavno usrednjavanje SMA (Simple moving average)
 - Kumulativno usrednjavanje CMA (Cumulative moving average)
 - Težinsko usrednjavanje WMA (Weigted moving average)

### Jednostavno usrednjavanje

$$SMA=\frac{P_m+P_{m-1}+P_{m-2}+...+P_{m-n}}{n} = \frac{1}{n}* \sum_{i=0}^{n-1} P_{m-i}$$ 


Na primeru merenja temperature ako postavimo n=10 onda bi rešenje izgledalo:

```python
n = 10

tavg = np.zeros(t.shape[0])
tavg[0:n] = temperature[0:n]

for i in range(10, t.shape[0]):
    tavg[i] = np.mean(temperature[i-n:i])
```

## Kumulativnim usrednjavanjem

$$CMA_n=\frac{P_1+...+P_n}{n} $$ 

$$CMA_{n+1}=\frac{P_{n+1}+...+n*CMA_n}{n+1} $$ 


```python
n = 10

tavg = np.zeros(t.shape[0])
tavg[0:n] = temperature[0:n]
tavg[n] = np.mean(tavg[0:n])

for i in range(n+1, t.shape[0]):
    tavg[i] = (temperature[i]+(i-1)*tavg[i-1])/i
```

```python
plt.plot(t, temperature, color='b', label='Temperatura')
plt.plot(t, tavg, color='r', label='CMA')
plt.legend()
plt.title('Merenje temperature')
plt.xlabel('vreme')
plt.ylabel('temperatura')
plt.grid(True)
```

## G-H filter

```python
def g_h_filter(data, x0, dx, g, h, dt=1.):
    x = x0
    results = []
    for z in data:
        #predikcija
        x_new = x + (dx*dt)

        #korekcija
        err = z - x_new
        dx = dx    + h * err / dt
        x  = x_new + g * err     
        results.append(x)  
    return np.array(results)
```

```python
tavg = g_h_filter(temperature, 30, 0.000, 0.2, 0.00002, 1)
```

```python
plt.plot(t, temperature, color='b', label='merenja')
plt.plot(t, tavg, color='r', label='g-h filter')
plt.legend()
plt.title('Merenje temperature')
plt.xlabel('vreme')
plt.ylabel('temperatura')
plt.grid(True)
```

## Kalman filter

```python
def predict(pos, movement):
    return (pos[0] + movement[0], pos[1] + movement[1])
```

```python
def gaussian_multiply(g1, g2):
    mu1, var1 = g1
    mu2, var2 = g2
    mean = (var1*mu2 + var2*mu1) / (var1 + var2)
    variance = (var1 * var2) / (var1 + var2)
    return (mean, variance)


def update(prior, likelihood):
    posterior = gaussian_multiply(likelihood, prior)
    return posterior
```

```python
process_var = 0.2 # variansa modela
sensor_var = 1. # variansa senzora

x = (-10., 10.)  # pocetna pretpostavka 

velocity = 0.003
process_model = (velocity*dt, process_var) 
```

```python
n = 5

tavg = np.zeros(t.shape[0])
tavg[0:n] = temperature[0:n]

for i in range(n, t.shape[0]):
    predikcija = predict(x, process_model)
    merenje = (temperature[i], sensor_var)
    x = update(predikcija, merenje)
    tavg[i] = x[0]
```

```python
plt.plot(t, temperature, color='b', label='merenja')
plt.plot(t, tavg, color='r', label='kalman filter')
plt.legend()
plt.title('Merenje temperature')
plt.xlabel('vreme')
plt.ylabel('temperatura')
plt.grid(True)
```
