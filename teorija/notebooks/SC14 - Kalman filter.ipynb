## Nepreciznost - neodređenost

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.linalg as linalg
%matplotlib inline
import pandas as pd
from matplotlib.patches import Circle

np.set_printoptions(precision=3)
```

```python
xp=2
yp=3

x = xp+np.random.normal(0, 1, 500)
y = yp+np.random.normal(0, 1, 500)

np.append(x, [200]);
np.append(y, [200]);
```

```python
circle1 = plt.Circle((2, 3), 3, color='g', alpha=0.2)
ax = plt.gca()
ax.add_artist(circle1)

plt.scatter(x, y, color='b', label='Merrenja');
plt.scatter(xp, yp, color='r', label='Tacno');
plt.legend()
plt.title('Normalna raspodela greske')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.grid(True)
```

```python
# udaljenost od tacne vrednosti
d = np.sqrt((x-xp)**2+(y-yp)**2)

plt.title('Histogram gresaka')
hist, bins = np.histogram(d, bins=10)
width = 0.7 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2
plt.bar(center, hist, align='center', width=width)
plt.show()
```
