```python
import dicom
import os
import numpy
from matplotlib import pyplot, cm
```

```python
#PathDicom = "d:\\Koljeno\\DICOM\\PA000000\\ST000000\\SE000001"
PathDicom = "images/dicom"
lstFilesDCM = []  # create an empty list
for dirName, subdirList, fileList in os.walk(PathDicom):
    for filename in fileList:
        lstFilesDCM.append(os.path.join(dirName,filename))

print lstFilesDCM[0]
```

```python
# Get ref file
ds = dicom.read_file(lstFilesDCM[0])

# Load dimensions based on the number of rows, columns, and slices (along the Z axis)
ConstPixelDims = (int(ds.Rows), int(ds.Columns))
print ConstPixelDims
ConstPixelSpacing = (float(ds.PixelSpacing[0]), float(ds.PixelSpacing[1]), float(ds.SliceThickness))

print ds.pixel_array[200:210, 200:210] 
```

```python
print numpy.amin(ds.pixel_array), numpy.amax(ds.pixel_array)
```

```python
x = numpy.arange(0.0, (ConstPixelDims[0]+1)*ConstPixelSpacing[0], ConstPixelSpacing[0])
y = numpy.arange(0.0, (ConstPixelDims[1]+1)*ConstPixelSpacing[1], ConstPixelSpacing[1])
```

```python
%matplotlib inline
pyplot.figure(dpi=300)
pyplot.axes().set_aspect('equal', 'datalim')
pyplot.set_cmap(pyplot.gray())
pyplot.pcolormesh(x, y, numpy.flipud(ds.pixel_array))
```
