```python
import math

# http://www.fundza.com/vectors/point2line/index.html
def dot(v,w):
    x,y = v
    X,Y = w
    return x*X + y*Y
  
def length(v):
    x,y = v
    return math.sqrt(x*x + y*y)
  
def vector(b,e):
    x,y = b
    X,Y = e
    return (X-x, Y-y)
  
def unit(v):
    x,y = v
    mag = length(v)
    return (x/mag, y/mag)
  
def distance(p0,p1):
    return length(vector(p0,p1))
  
def scale(v,sc):
    x,y = v
    return (x * sc, y * sc)
  
def add(v,w):
    x,y = v
    X,Y = w
    return (x+X, y+Y)
```

Projekcija tačke na liniju

```python
def pnt2line(pnt, start, end):
    line_vec = vector(start, end)
    pnt_vec = vector(start, pnt)
    line_len = length(line_vec)
    line_unitvec = unit(line_vec)
    pnt_vec_scaled = scale(pnt_vec, 1.0/line_len)
    t = dot(line_unitvec, pnt_vec_scaled)    
    if t < 0.0:
        t = 0.0
    elif t > 1.0:
        t = 1.0
    nearest = scale(line_vec, t)
    dist = distance(nearest, pnt_vec)
    nearest = add(nearest, start)
    return (dist, (int(nearest[0]), int(nearest[1])))
```

```python
import matplotlib.pyplot as plt  # za prikaz slika, grafika, itd.
%matplotlib inline
import cv2
import numpy as np
import random
import time
from scipy import ndimage
import math
import matplotlib.pyplot as plt
```

```python
width = 640
height = 480

objs = []
line = [(100,450), (500, 100)]

random.seed(123)
```

```python
def init():
    for i in range(0, 30):
        xc = random.randint(20, width)
        yc = random.randint(20, height)
        objs.append({'x':xc, 'y':yc, 'pass':False})
```

```python
def generate_image():
    img = np.zeros((height,width,3), np.uint8)
    for i in range(len(objs)):
        obj = objs[i]
        obj['x'] += 1
        if i%2==0:
            obj['y'] += 1
        xc = obj['x']
        yc = obj['y']
        
        dist, pnt = pnt2line((xc, yc), line[0], line[1])
        cv2.line(img, pnt, (xc, yc), (0, 255, 25), 1)
        c = (25, 25, 255)
        if(dist<5):
            c = (0, 255, 160)
            if obj['pass'] == False:
                obj['pass'] = True
        cv2.circle(img, (xc,yc), 16, c, 2)

    cv2.line(img, line[0], line[1] , (255, 0, 0), 3)
    return img
```

```python
init()

img = generate_image()
plt.imshow(img)
```
